### 글 목록 link list 와 댓글 가져올 수 있는 function 만들기 

```python
def getLink(url, params=None, select=None, num_retries=2):
    html = requests.get(url, params=params, headers=headers)
    
    if 500<=html.status_code<600 and num_retries>0:
        print(html.status_code, html.reason)
        return getUrl(url=url, params=params, num_retries = (num_retries - 1))
    
    
    #domtree로 만들기
    dom = bs(html.text, "html.parser")
    links = dom.select(select)
    
    return [link["href"] for link in links if link.has_attr("href") == True]
```

```python
def getContent(url, params=None, select=None, num_retries=2):
    html = requests.get(url, params=params, headers=headers)
    
    if 500<=html.status_code<600 and num_retries>0:
        print(html.status_code, html.reason)
        return getContent(url=url, params=params, num_retries = (num_retries - 1))
    
    dom = bs(html.text, "html.parser")
    content = dom.select(select)
    
    return content
```

```python
def getRipple(url, params=None, select=None, num_retries=2):
    html = requests.get(url, params=params, headers=headers)
    
    if 500<=html.status_code<600 and num_retries>0:
        print(html.status_code, html.reason)
        return getReple(url=url, params=params, num_retries = (num_retries - 1))
    
    dom = bs(html.text, "html.parser")
    ripples = dom.select(select)
    
    return [ripple for ripple in ripples]
```
